# 랜덤 숫자

```
# random

import numpy as np
np.random.rand(3) # rand -> 0~1 사이의 랜덤 숫자 출력
```

```
# 몬테카를로 분석 (원주율 구하기)

import numpy as np
import matplotlib.pyplot as plt

n= 10000
x= np.random.rand(n)
y= np.random.rand(n)

plt.figure(figsize= (6, 6)) # plt.figure -> 새로운 figure 생성, figsize -> figure의 크기 설정
plt.scatter(x, y, s= 1) # s -> 점의 크기
xx= np.linspace(0, 1, 100) # 0과 1 사이에 100개의 일정한 간격의 배열 생성
plt.plot(xx, (1-xx*xx)**0.5, c= 'r') # c -> 색깔

pi= ((x**2 + y**2) < 1).mean()*4
pi
```

```
# 정규 분포

np.random.randint(1, 7, 10) # 1이상 7미만의 정수 중에 10개 랜덤으로 뽑기
```

```
x = np.random.randint(1, 7, 10000)
plt.hist(x, bins= 6, rwidth= 0.8) # bins -> 막대가 6개, rwidth -> 막대의 너비
plt.show()
```

```
z = 0
n = 2 # 주사위 던진 개수
for i in range(n):
  x = np.random.randint(1, 7, 10000)
  z = z + x

plt.hist(z, bins= n*5+1, rwidth= 0.6)
plt.show()
```
