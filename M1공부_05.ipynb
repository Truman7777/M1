```
# 라이브러리 불러오기
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from pandas import DataFrame, Series
from numpy.random import randn
%matplotlib inline
```

```
fig = plt.figure()
ax1 = fig.add_subplot(2, 2, 1) # add_subplot -> figure에 격자 나누기
ax2 = fig.add_subplot(2, 2, 2)
ax3 = fig.add_subplot(2, 2, 3)
```

```
ax1.hist(np.random.randn(200), bins = 20, color= 'k', rwidth= 0.5)
ax2.scatter(np.arange(30), np.arange(30) + 3*np.random.randn(30))
ax3.plot(np.arange(10), np.random.randn(10))

fig
```

```
fig, axes = plt.subplots(2, 2, sharex= True, sharey= True) # sharex= True -> x축 공유, sharey= True -> y축 공유
for i in range(2):
  for j in range(2):
    axes[i, j].hist(np.random.randn(500),
                    bins= 50, color= 'k', alpha= 0.6) # alpha -> 투명도(1로 갈수록 짙음)
plt.subplots_adjust(wspace= 0, hspace= 0) # wspace -> subplot 간의 좌우간격, hspace -> subplot 간의 상하간격
```

```
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)
ax.plot(randn(1000).cumsum(), 'k', label= 'one') # cumsum -> 배열에서 누적되는 원소들의 합을 계산
ax.plot(randn(1000).cumsum(), 'k--', label= 'two')
ax.plot(randn(1000).cumsum(), 'k.', label= 'three')

ax.legend(loc= 'best') # loc= 'best' -> label을 최적의 위치에 표시함
```

```
fig, axes = plt.subplots(2, 1)
data = Series(np.random.rand(16),
              index= list('abcdefghijklmnop'))
data.plot(kind= 'bar', ax= axes[0], color= 'k', alpha= 0.9) # bar -> 세로 막대 그래프
data.plot(kind= 'barh', ax= axes[1], color= 'k', alpha= 0.7) # barh -> 가로 막대 그래프
```

```
fig = plt.figure()
ax = fig.add_subplot(1, 1, 1)

rect = plt.Rectangle((0.2, 0.75), 0.5, 0.2, color= 'k', alpha= 0.3) # 시작점과 가로길이, 세로길이
circ = plt.Circle((0.7, 0.2), 0.15, color= 'b', alpha= 0.3) # 원의 중심 좌표와 반지름 길이
pgon = plt.Polygon([[0.15, 0.15], [0.35, 0.4], [0.2, 0.6]], color= 'g', alpha= 0.5) # 꼭짓점 좌표를 입력해 도형 만들기

ax.add_patch(rect) # ax subplot에 도형 추가하기
ax.add_patch(circ)
ax.add_patch(pgon)
```

```
df = DataFrame(np.random.rand(6, 4),
               index= ['one', 'two', 'three', 'four', 'five', 'six'],  columns= pd.Index(['A', 'B', 'C', 'D'], 
                                                                                         name= 'Genus'))
df.plot(kind= 'bar')
```

```
df.plot(kind= 'barh', stacked= True) # stacked= True -> 하나의 인덱스에 대한 각 열의 값을 한줄로 쌓아서 나타냄
```

```
comp1 = np.random.normal(0, 1, size= 200) # 평균0, 표준편차1인 정규분포로부터 200개 무작위 추출
comp2 = np.random.normal(10, 2, size= 200) #평균10, 표준편차2인 정규분포로부터 200개 무작위 추출
values = Series(np.concatenate([comp1, comp2])) # np.concatenate -> 배열 합치기
values.hist(bins= 100, alpha= 0.3, color= 'k', density= True)
values.plot(kind= 'kde', style= 'k--') # 'kde' -> 밀도 추정
```
